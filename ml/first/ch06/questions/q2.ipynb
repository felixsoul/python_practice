```python
import math 
import random
import matplotlib.pyplot as plt
num = 100
x_start = -3
x_end = 3
step = (x_end - x_start) / (num - 1)
x_data = [-3 + i * step for i in range(num)]
y_data = [math.exp(x_idx) for x_idx in x_data]
fig, ax = plt.subplots()
ax.spines['left'].set_position('zero')  # 左边轴交叉在原点
ax.spines['bottom'].set_position('zero')  # 底部轴交叉在原点
ax.spines['right'].set_color('none')  # 右边轴隐藏
ax.spines['top'].set_color('none') 
ax.plot(x_data, y_data)
ax.set_xlabel('x');ax.set_ylabel('y')
#ax.set_aspect('equal', adjustable='box')
y_data_2 = [math.exp(-x_idx ** 2) for x_idx in x_data]
ax.plot(x_data, y_data_2, color="red")
ax.grid()
```
