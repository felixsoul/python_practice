```python
import random
import statistics
import matplotlib.pyplot as plt

#抛色子实验的次数
num_flips = 1000

#模拟抛色子实验
#用于存储每次抛色子的结果
results = [random.choices([1,2,3,4,5,6],[0.2, 0.16, 0.16, 0.16, 0.16, 0.16])[0] for _ in range(num_flips)]
running_means = [statistics.mean(results[0:idx+1]) for idx in range(num_flips)]

#可视化前100次结果均值随次数变化
visual_num = 100
plt.scatter(range(1, visual_num+1), results[0:visual_num],
            c = results[0:visual_num], marker="o", cmap='cool')
plt.plot(range(1, visual_num+1), results[0:visual_num])
plt.plot(range(1, num_flips + 1), running_means)
plt.axhline(3.5, color='r')
plt.xlabel("Number of coin flips")
plt.ylabel("Running Mean")
plt.grid(True)
plt.show()
```
